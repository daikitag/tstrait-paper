```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2,1,figsize=(7,10))

num_ind = [1000, 5000, 10_000, 25_000, 50_000, 100_000, 250_000, 500_000,
          1_000_000, 2_000_000, 3_000_000, 4_000_000, 5_000_000]
fifty = [0.231, 0.443, 0.697, 1.211, 2.731, 3.748, 14.91, 37.358,
        78.383, 175.031, 252.927, 346.200, 450.884]
hundred = [0.353, 0.606, 0.85, 1.437, 2.662, 4.934, 17.051, 52.748,
          74.525, 165.859, 253.093, 337.865, 418.403]
two_hundred = [0.608, 0.99, 1.297, 2.01, 3.263, 7.199, 14.799, 36.684,
              82.211, 165.803, 257.556, 363.875, 435.369]

previous_num_ind = [1000, 5000, 10_000, 25_000, 50_000, 100_000, 250_000]
previous = [3.982, 9.517, 16.282, 45.325, 92.817, 210.099, 1000.41]

axs[0].plot(num_ind, fifty, label="tstrait: 50 Mb", marker=".")
axs[0].plot(num_ind, hundred, label="tstrait: 100 Mb", marker=".")
axs[0].plot(num_ind, two_hundred, label="tstrait: 200 Mb", marker=".")
axs[0].plot(previous_num_ind, previous, label="Martin et al. (2017): 100 Mb", marker=".")
axs[0].set_xlabel('Number of Individuals')
axs[0].set_ylabel('Time (Seconds)')
axs[0].text(100, 450, "A", weight="bold", size=20)
axs[0].set_ylim((0,500))
axs[0].legend(loc="lower right")

num_causal = [50, 100, 500, 1000, 3000, 5000, 7000, 10000]
french = [34.991, 37.664, 50.979, 60.845, 98.623, 155.9, 209.674, 265.691]
thousand = [3.001, 2.579, 2.552, 2.93, 3.698, 3.822, 4.367, 5.943]
axs[1].loglog(num_causal, french, label="French Canadian", marker=".")
axs[1].loglog(num_causal, thousand, label="1000 Genomes Project", marker=".")
axs[1].set_xlabel('Number of Causal Sites (log scale)')
axs[1].set_ylabel('Time (Seconds, log scale)')
axs[1].legend(loc="center right")
plt.text(50, 140, "B", weight="bold", size=20)

plt.show()
```
